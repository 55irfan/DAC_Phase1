```python
import pandas as pd

# Load the dataset from the CSV file
df = pd.read_csv('Survey_dataset.csv')

# Replace 'NA' with NaN for missing values
df.replace('NA', pd.NA, inplace=True)

# Handle missing values, e.g., fill NaN values with appropriate values
# For numerical columns, you can use methods like fillna()
# For categorical columns, you can use fillna() or mode()

# Example: Filling missing values in 'self_employed' with the mode
df['self_employed'].fillna(df['self_employed'].mode()[0], inplace=True)

# Convert 'Age' to integers
df['Age'] = df['Age'].astype(int)

# Handle categorical data by encoding them (if needed)
# You can use methods like get_dummies() for one-hot encoding

# Example: One-hot encoding 'Gender' column
df = pd.get_dummies(df, columns=['Gender'], prefix='Gender')

# Remove any unnecessary columns
# For example, if a column is mostly missing, you can drop it
df.drop('comments', axis=1, inplace=True)

# You can perform similar operations for other columns as needed

# Verify the cleaned dataset
print(df.head())

# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_dataset.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv('cleaned_dataset.csv')

# Visualize the distribution of 'Age'
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()



# Visualize the relationship between 'remote_work' and 'treatment'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='remote_work', hue='treatment')
plt.title('Remote Work vs. Treatment')
plt.xlabel('Remote Work')
plt.ylabel('Count')
plt.show()

# Visualize the relationship between 'mental_health_consequence' and 'phys_health_consequence'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='mental_health_consequence', hue='phys_health_consequence')
plt.title('Mental Health Consequence vs. Physical Health Consequence')
plt.xlabel('Mental Health Consequence')
plt.ylabel('Count')
plt.show()
```
